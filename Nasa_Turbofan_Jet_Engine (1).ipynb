```python
import pandas as pd
data_train = pd.read_csv('/content/CMaps/train_FD001.txt',sep=' ',header=None)
data_test = pd.read_csv('/content/CMaps/test_FD001.txt',sep=' ',header=None)
```

```python
index_names = ['engine', 'cycle']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names=[ "(Fan inlet temperature) (â—¦R)",
"(LPC outlet temperature) (â—¦R)",
"(HPC outlet temperature) (â—¦R)",
"(LPT outlet temperature) (â—¦R)",
"(Fan inlet Pressure) (psia)",
"(bypass-duct pressure) (psia)",
"(HPC outlet pressure) (psia)",
"(Physical fan speed) (rpm)",
"(Physical core speed) (rpm)",
"(Engine pressure ratio(P50/P2)",
"(HPC outlet Static pressure) (psia)",
"(Ratio of fuel flow to Ps30) (pps/psia)",
"(Corrected fan speed) (rpm)",
"(Corrected core speed) (rpm)",
"(Bypass Ratio) ",
"(Burner fuel-air ratio)",
"(Bleed Enthalpy)",
"(Required fan speed)",
"(Required fan conversion speed)",
"(High-pressure turbines Cool air flow)",
"(Low-pressure turbines Cool air flow)" ]
col_names = index_names + setting_names + sensor_names
```

```python
data_train.columns = col_names
data_train.describe()

data_test.columns = col_names
data_test.describe()
```

```python
# Define a function to calculate the remaining useful life (RUL)
def add_rul(g):
    # Calculate the RUL as the difference between the maximum cycle value and the cycle value for each row
    g['RUL'] = max(g['cycle']) - g['cycle']
    return g

# Apply the add_rul function to the training data grouped by the unit ID
data_train = data_train.groupby('engine').apply(add_rul)
data_test = data_test.groupby('engine').apply(add_rul)
```

```python
import torch
from torch import nn
```

```python
x_train = data_train.iloc[:,0:26]
x_train = torch.Tensor(x_train.to_numpy()).to(device)
```

```python
y_train = data_train.iloc[:,26:]
y_train = torch.Tensor(y_train.to_numpy()).to(device)
```

```python
x_test = data_train.iloc[:,0:26]
x_test = torch.Tensor(x_test.to_numpy()).to(device)

y_test = data_train.iloc[:,26:]
y_test = torch.Tensor(y_test.to_numpy()).to(device)
```

```python
#Model building
#class model_0(nn.Module):
#  def __init__(self):
#      super().__init__()
#      self.linear_layer = nn.Sequential(
#          nn.Linear(in_features= 26,out_features= 100),
#          nn.Linear(in_features= 100,out_features= 200),
#          nn.Linear(in_features= 200,out_features= 1))

#      def forward(self,x:torch.Tensor)->torch.Tensor:
#        return linear_layer(x)
```

```python
model_0 = nn.Sequential(
          nn.Linear(in_features= 26,out_features= 100),
         #nn.ReLU(),
          nn.Linear(in_features= 100,out_features= 300),
          #nn.ReLU(),
          nn.Linear(in_features = 300,out_features = 500),
          #nn.ReLU(),
          nn.Linear(in_features= 500,out_features= 1)).to(device)
```

```python
#Set up loss function and optimizer
loss_fn = nn.L1Loss()

optimizer = torch.optim.SGD(params = model_0.parameters(), lr = 0.01)
```

```python
def accuracy_score(actual, predicted):
 correct = 0
 for i in range(len(actual)):
  if actual[i] == predicted[i]:
    correct += 1
 return correct / float(len(actual)) * 100.0
```

```python
#Training loop
RANDOM_SEED = 42
torch.manual_seed(42)

epochs = 1000
i = 0
for epoch in range(epochs):
  model_0.train()

  y_pred = model_0(x_train).to(device)
  print(y_pred)

  loss = loss_fn(y_pred,y_train)

  optimizer.zero_grad()

  loss.backward()

  optimizer.step()

  i=i+1

  model_0.eval() #Testing phase

  with torch.inference_mode():
    test_pred = model_0(x_test).to(device)
    #print(test_pred)
    #test_loss = loss_fn(test_pred,y_pred)

  if i%100 == 0:
    acc = accuracy_score(y_test,test_pred)
    print("Accuracy for "+ str(i) + " is " + str(acc))

```
