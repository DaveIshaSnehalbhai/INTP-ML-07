```python
import pandas as pd
data_train = pd.read_csv('/content/CMaps/train_FD001.txt',sep=' ',header=None)
data_test = pd.read_csv('/content/CMaps/test_FD001.txt',sep=' ',header=None)
```

```python
data_train.drop(columns=data_train.columns[-1],  axis=1,  inplace=True)
data_test.drop(columns = data_test.columns[-1],  axis=1,  inplace=True)

print(data_test)
```

```python
data_test.drop(data_test.columns[26],  axis=1, inplace=True)
data_test
```

```python
index_names = ['engine', 'cycle']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names=[ "(Fan inlet temperature) (◦R)",
"(LPC outlet temperature) (◦R)",
"(HPC outlet temperature) (◦R)",
"(LPT outlet temperature) (◦R)",
"(Fan inlet Pressure) (psia)",
"(bypass-duct pressure) (psia)",
"(HPC outlet pressure) (psia)",
"(Physical fan speed) (rpm)",
"(Physical core speed) (rpm)",
"(Engine pressure ratio(P50/P2)",
"(HPC outlet Static pressure) (psia)",
"(Ratio of fuel flow to Ps30) (pps/psia)",
"(Corrected fan speed) (rpm)",
"(Corrected core speed) (rpm)",
"(Bypass Ratio) ",
"(Burner fuel-air ratio)",
"(Bleed Enthalpy)",
"(Required fan speed)",
"(Required fan conversion speed)",
"(High-pressure turbines Cool air flow)",
"(Low-pressure turbines Cool air flow)" ]
col_names = index_names + setting_names + sensor_names
```

```python
data_train.columns = col_names
data_train.describe()

data_test.columns = col_names
data_test.describe()
```

```python
# Define a function to calculate the remaining useful life (RUL)
def add_rul(g):
    # Calculate the RUL as the difference between the maximum cycle value and the cycle value for each row
    g['RUL'] = max(g['cycle']) - g['cycle']
    return g

# Apply the add_rul function to the training data grouped by the unit ID
data_train = data_train.groupby('engine').apply(add_rul)
data_test = data_test.groupby('engine').apply(add_rul)
```

```python
print(data_train.shape)
data_test.shape
```

```python
import torch
from torch import nn
```

```python
x_train = data_train.iloc[:,0:26].values
x_train = torch.Tensor(x_train).to(device)
x_train.shape
x_train
```

```python
y_train = data_train.iloc[:,26:].values
y_train = torch.Tensor(y_train).to(device)
y_train.shape
y_train
```

```python
x_test = data_test.iloc[:,0:26].values
x_test = torch.Tensor(x_test).to(device)

y_test = data_test.iloc[:,26:].values
y_test = torch.Tensor(y_test).to(device)

print(x_test.shape)
print(y_test.shape)
x_test.type
```

```python
#Model building
class model_0(nn.Module):
  def __init__(self):
      super().__init__()
      self.linear_layer = nn.Sequential(
          nn.Linear(in_features= 26,out_features= 100),
          nn.Linear(in_features= 100,out_features= 200),
          nn.Linear(in_features= 200,out_features= 1))

  def forward(self,x):
      return self.linear_layer(x)
```

```python
#model_0 = nn.Sequential(
#          nn.Linear(in_features= 26,out_features= 100),
         #nn.ReLU(),
#          nn.Linear(in_features= 100,out_features= 300),
          #nn.ReLU(),
#          nn.Linear(in_features = 300,out_features = 500),
          #nn.ReLU(),
#          nn.Linear(in_features= 500,out_features= 1)).to(device)
```

```python
torch.manual_seed(42)

model = model_0().to(device)
with torch.inference_mode():
  y_preds = model(x_test).to(device)
y_preds
```

```python
#Set up loss function and optimizer
loss_fn = nn.BCEWithLogitsLoss()

optimizer = torch.optim.Adam(params = model.parameters(), lr = 0.1)
```

```python
def accuracy_score(actual, predicted):
 correct = 0
 for i in range(len(actual)):
  if actual[i] == predicted[i]:
    correct += 1
 return correct / float(len(actual)) * 100.0
```

```python
#Training loop
torch.manual_seed(42)

epochs = 1000
i = 0
for epoch in range(epochs):
  model.train()

  y_logits = model(x_train)
  y_pred = torch.round(torch.sigmoid(y_logits))
  #print(y_pred)

  loss = loss_fn(y_logits,y_train)
  #print(loss)

  if i%100 ==0:
    acc = accuracy_score(y_train,y_pred)
    print("Training acc ",i,"Is ",acc)

  optimizer.zero_grad()

  loss.backward()

  optimizer.step()

  i=i+1

  model.eval() #Testing phase

  with torch.inference_mode():
    test_logits = model(x_test)
    test_pred = torch.round(torch.sigmoid(test_logits))
    #print(test_pred)
    test_loss = loss_fn(test_logits,y_test)

  if i%100 == 0:
    acc = accuracy_score(y_test,test_pred)
    print("Testing acc ",i,"Is ",acc)


```

```python
#Linear Regression Model
X = data_train.iloc[:,0:26]
Y = data_train.iloc[:,26:]

#Train and test data splitting

train_split = int(0.8* len(data_train))
x_train,y_train = X[:train_split],Y[:train_split]
x_test,y_test = X[train_split:],Y[train_split:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import  mean_squared_error,r2_score, mean_absolute_percentage_error
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
import numpy as np
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean squared error: {mse}")
print(f"Root mean squared error: {rmse}")
print(f"R-squared score: {r2}")
print(f"mean_absolute_percentage_error: {mape}")
```

```python
training_score = cross_val_score(model, x_train, y_train, cv=5)
print("Algorithm: ", model, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```
